```python
from diffusers import DiffusionPipeline
import torch

pipeline = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, use_safetensors=True, variant="fp16")
pipeline.load_lora_weights("blink7630/storyboard-sketch")
pipeline.to("cuda")

prompt = "An astronaut riding a green horse" # Your prompt here
neg_prompt = "ugly, blurry, poor quality" # Negative prompt here
image = pipeline(prompt=prompt, negative_prompt=neg_prompt).images[0]
```
